## Import of libs

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading data

```python
lifexpectancy_database = pd.read_csv('databases/lifeExpectancy_world.csv')
```

```python
lifexpectancy_database
```

## Best and worst expectations

## Worst expectation

```python
years_database = lifexpectancy_database.iloc[:, 2:72]
```

```python
years_database
```

```python
minimum = years_database.iloc[:, 0:70].min().min()
```

```python
minval_col = years_database.min().idxmin()
minval_row = years_database.min(axis='columns').idxmin()
```

```python
minval_row, minval_col
```

```python
minimum_dataset = lifexpectancy_database[[minval_col]].loc[[minval_row]]
minimum_dataset
```

```python
minimum_completdataset = lifexpectancy_database.loc[lifexpectancy_database['1977'] == minimum]
minimum_completdataset
```

```python
sns.catplot(data = minimum_completdataset).set_xlabels('1950-2019', weight='bold').set(xticklabels=[], title='Cambodia').set_ylabels('Life expectancy', weight='bold');
plt.text(15, 18, '1977 -->', fontweight='bold');
```

```python
minimum
```

```python
sns.catplot(x = '1977', data = lifexpectancy_database).set_xlabels('MIN-MAX', fontweight='bold').set(xticklabels=[], title='Cambodia compared to other places');
plt.text(20, 0.1, 'Cambodia', fontweight='bold');
plt.text(43, -0.45, '1977', fontweight='bold');
```

```python
maximumame_dataset = lifexpectancy_database[[maxvalame_col]].loc[[maxvalame_row]]
maximumame_dataset
```

```python
sns.catplot(data = america_dataset).set_xlabels('1950-2019', weight='bold').set(xticklabels=[], title='America').set_ylabels('Life expectancy', weight='bold');
plt.text(57.3, 76.8, '2019 -->', fontweight='bold');
plt.text(2.6, 59, '<-- 1950', fontweight='bold');
```

```python
asia_dataset = lifexpectancy_database[lifexpectancy_database['Entity'] == 'Asia']
```

```python
africa_dataset = lifexpectancy_database[lifexpectancy_database['Entity'] == 'Africa']
```

```python
europe_dataset = lifexpectancy_database[lifexpectancy_database['Entity'] == 'Europe']
```

```python
laameandcarribean_dataset = lifexpectancy_database[lifexpectancy_database['Entity'] == 'Latin America and the Caribbean']
```

```python
oceania_dataset = lifexpectancy_database[lifexpectancy_database['Entity'] == 'Oceania']
```

```python
fig, ax = plt.subplots(1,6, figsize=(20, 5))

ax2 = sns.stripplot(data = america_dataset, ax=ax[0])

ax2.set(xticklabels=[], title='America')
ax2.set_xlabel('1950-2019', weight='bold')
ax2.set_ylabel('Life expectancy', weight='bold')


ax3 = sns.stripplot(data = asia_dataset, ax=ax[1])

ax3.set(xticklabels=[], title='Asia');
ax3.set_xlabel('1950-2019', weight='bold')
ax3.set_ylabel('Life expectancy', weight='bold')

ax4 = sns.stripplot(data = africa_dataset, ax=ax[2])
ax4.set(xticklabels=[], title='Africa')
ax4.set_xlabel('1950-2019', weight='bold')
ax4.set_ylabel('Life expectancy', weight='bold')

ax5 = sns.stripplot(data = europe_dataset, ax=ax[3])
ax5.set(xticklabels=[], title='Europe')
ax5.set_xlabel('1950-2019', weight='bold')
ax5.set_ylabel('Life expectancy', weight='bold')

ax6 = sns.stripplot(data = laameandcarribean_dataset, ax=ax[4])
ax6.set(xticklabels=[], title='Latin America and Caribbean')
ax6.set_xlabel('1950-2019', weight='bold')
ax6.set_ylabel('Life expectancy', weight='bold')

ax6 = sns.stripplot(data = oceania_dataset, ax=ax[5])
ax6.set(xticklabels=[], title='Oceania')
ax6.set_xlabel('1950-2019', weight='bold')
ax6.set_ylabel('Life expectancy', weight='bold')

fig.tight_layout()
```
